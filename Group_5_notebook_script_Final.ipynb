```python
# Importing necessary libraries to load the datasets
import pandas as pd
```

```python
# Loading the dataset for incoming shoes
df = pd.read_excel("/content/Incoming_shoes_from_partners.xlsx")
df
```

```python
df.info() # Check the features data types
```

```python
df.columns # Checking the feature names
```

```python
df.drop(index=5, inplace=True) # Dropping useless row
```

```python
df['Date/Year'] = pd.to_datetime(df['Date/Year']).dt.date # Converting Date/Year from object to date
df
```

```python
# Importing libraries to create a chart
import seaborn as sns
import matplotlib.pyplot as plt

# Making the line chart and assigning the x and y variables
sns.lineplot(data=df, x='Date/Year', y='Amount of shoes', marker='o')

# Adding labels to the chart
plt.title('Amount of shoes received')
plt.xlabel('Date')
plt.ylabel('Amount of shoes')
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
df.to_csv("Cleaned_incoming_shoes.csv")
```
